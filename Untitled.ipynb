Virtual data creation

```python
import pandas as pd
import random

#data pools
locations = ['Ahmedabad', 'Surat', 'Mumbai', 'Pune', 'Delhi']
colleges = ['XYZ University', 'ABC Institute', 'PQR College', 'DEF Academy']
years_of_study = ['1st', '2nd', '3rd', '4th']
program_interests = ['Data Science', 'Robotics', 'AI', 'Electric Vehicle']
lead_sources = ['Instagram', 'LinkedIn', 'College Collaboration', 'Google Form', 'Mass-Mailing', 'Whatsapp']

# Generating dataset
data = {
    'Lead ID': [f'LD{i}' for i in range(1, 10001)],
    'Location': [random.choice(locations) for _ in range(10000)],
    'College': [random.choice(colleges) for _ in range(10000)],
    'Year of Study': [random.choice(years_of_study) for _ in range(10000)],
    'Program Interest': [random.choice(program_interests) for _ in range(10000)],
    'Lead Source': [random.choice(lead_sources) for _ in range(10000)],
}

# DataFrame
df = pd.DataFrame(data)
df.to_csv('lead_dataset.csv', index=False)
```

Demographic Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('lead_dataset.csv')

# Location Analysis
location_counts = df['Location'].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=location_counts.index, y=location_counts.values, palette="viridis")
plt.title('Leads by Location')
plt.ylabel('Number of Leads')
plt.xlabel('Location')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# College Analysis
college_counts = df['College'].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=college_counts.index, y=college_counts.values, palette="magma")
plt.title('Leads by College')
plt.ylabel('Number of Leads')
plt.xlabel('College')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Year of Study Analysis
year_counts = df['Year of Study'].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=year_counts.index, y=year_counts.values, palette="plasma")
plt.title('Leads by Year of Study')
plt.ylabel('Number of Leads')
plt.xlabel('Year of Study')
plt.tight_layout()
plt.show()
```

Program Analysis

```python
# Program Interest Analysis
program_counts = df['Program Interest'].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=program_counts.index, y=program_counts.values, palette="cool")
plt.title('Program Interest Levels')
plt.ylabel('Number of Leads')
plt.xlabel('Program Interest')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Interest by Location (Heatmap)
program_location = pd.crosstab(df['Program Interest'], df['Location'])
plt.figure(figsize=(10, 6))
sns.heatmap(program_location, cmap="YlGnBu", annot=True, fmt='d')
plt.title('Program Interest by Location')
plt.ylabel('Program Interest')
plt.xlabel('Location')
plt.tight_layout()
plt.show()
```

Projections and Budget Allocation

```python
# Simulate conversion rates (10-20%)
df['Conversion Rate'] = [random.uniform(0.1, 0.2) for _ in range(len(df))]
df['Converted Leads'] = (df['Conversion Rate'] * 100).astype(int)

# Summarize conversions by program
conversion_summary = df.groupby('Program Interest')['Converted Leads'].sum().sort_values(ascending=False)

# Budget allocation suggestion (proportional to leads converted)
total_converted = conversion_summary.sum()
conversion_summary = conversion_summary.reset_index()
conversion_summary['Budget (%)'] = (conversion_summary['Converted Leads'] / total_converted * 100).round(2)

print(conversion_summary)
```

Report Creation

```python
# Export visuals and summary to a file
conversion_summary.to_csv('conversion_summary.csv', index=False)

# Save plots as images
location_plot_path = 'location_analysis.png'
plt.figure()
sns.barplot(x=location_counts.index, y=location_counts.values, palette="viridis")
plt.title('Leads by Location')
plt.savefig(location_plot_path)

print("All results and visualizations are saved.")
```
